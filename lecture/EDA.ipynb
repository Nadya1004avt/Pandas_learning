Создайте столбец weekday, в котором будет храниться день недели в русском варианте (Понедельник, Вторник, ...)
Создайте столбец date, в котором будет храниться дата регистрации пользователя (без времени)

```python
df1 = df.copy()
```

```python
week_rus = {'Monday': 'Понедельник',
           'Tuesday': 'Вторник',
           'Wednesday': 'Среда',
           'Thursday': 'Четверг',
           'Friday': 'Пятница',
           'Saturday': 'Суббота',
           'Sunday': 'Воскресенье'}

df1['weekday'] = pd.to_datetime(df1['date_joined']).dt.day_name()
df1['weekday'] = df1['weekday'].map(week_rus)
df1['date'] = pd.to_datetime(df['date_joined']).dt.date
```

```python
df1
```

Создайте датафрейм res, который получится в результате группировки исходного датафрейма по дням недели и расчета:
Количества регистраций в этот день недели
Количества уникальных дат, в которые были регистрации для этого дня недели

```python
output = df1.groupby('weekday', as_index=False).agg(
    count=('date', 'count'),
    nunique=('date', 'nunique')
)
output
```

Отфильтруйте только тех пользователей, у которых id >= 94. Пользователи с меньшим id - фейковые, они сбивают статистику.
2. Также уберите пользователей с одним из этих ников: 'fnoxonsnbkbxsnb', 'Jonns21241412', 'fbxbnsn'. Это также фейковые юзеры.

```python
df2 = df.copy()
df2 = df2[(df2['id'] >= 94) & ~df2['username'].isin(['fnoxonsnbkbxsnb', 'Jonns21241412', 'fbxbnsn'])]
df2
```

Создайте столбец cohort, который будет содержать год и месяц регистрации в формате YYYY-MM.

```python
df2['cohort'] = pd.to_datetime(df['date_joined']).dt.strftime('%Y-%m')
df2
```

Создайте переменную res, которая будет содержать долю каждой когорты в общем количестве зарегистрированных пользователей. В сумме все доли должны давать 100.

```python
res = df2.groupby('cohort')['id'].count()
total_users = res.sum()
res = round((res / total_users)*100, 6)
res.name = 'proportion'
res
```

```python
coderun = pd.read_csv('../data/itresume-coderun.csv',encoding='1251')
coderun
```

Создайте столбец cohort, который будет содержать год и месяц регистрации в формате YYYY-MM. 
Создайте столбец language, который проводит соответствие между языком и language_id

```python
coderun1 = coderun.copy()
lang = {
    3 : 'Python',
    2 : 'SQL'
}
coderun1['cohort'] = pd.to_datetime(coderun1['created_at']).dt.strftime('%Y-%m')
coderun1['language'] = coderun1['language_id'].map(lang)
coderun1 = coderun1[coderun1['cohort'] >= '2022-01']

coderun1
```

А теперь давайте посчитаем - в скольких задачах в среднем один пользователь отправлял на выполнение код в каждой когорте по каждому языку.

```python
coderun1 = coderun1.groupby(['cohort', 'language', 'user_id']).agg(task_count = ('problem_id', 'nunique')).reset_index().groupby(['cohort', 'language']).agg(avg_tasks = ('task_count', 'mean')).reset_index()
coderun1
```

```python
df3 = coderun1.pivot_table(index = 'cohort', columns = 'language', values = 'avg_tasks')
df3
```

language_id - номер языка, который нужно оставить. Значение: 2.
