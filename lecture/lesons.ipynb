Соединение датафреймов

```python
abc = pd.DataFrame({
	'name' : ['кабачок', 'помидор', 'огурец'],
	'abc' : ['A', 'A', 'C']
})

xyz = pd.DataFrame({
	'name_xyz' : ['кабачок', 'баклажан', 'огурец'],
	'xyz' : ['X', 'Y', 'Z']
})
```

```python
abc_xyz = abc.merge(xyz, 
		  left_on = 'name',
		  right_on = 'name_xyz',
		  how = 'outer'
		  )

abc_xyz
```

```python
abc_xyz['name'] = np.where(abc_xyz['name'].isnull(), abc_xyz['name_xyz'], abc_xyz['name'])

abc_xyz
```

axis = 1 удаление по столбцам. По умолчанию удаление по строкам. 

```python
abc_xyz.drop('name_xyz', axis = 1)
```

Разбивание датафрейма и сохранение в файлы

```python
shops = df['DR_Apt'].unique().tolist()
shops
```

```python
for shop in shops:
	small_df = df[df['DR_Apt'] == shop]
	small_df.to_excel(f'extra/{shop}.xlsx', sheet_name = f'{shop}', index = False)
```

Собрать в один датафрейм отельные файлы excel

```python
glob.glob('extra/*.xlsx')
```

```python
files = glob.glob('extra/*.xlsx')
dfs = []

for file_name in files:
	small_df = pd.read_excel(file_name)
	dfs.append(small_df)
```

```python
len(dfs)
```

```python
df = pd.concat(dfs, ignore_index = True)
df
```

Пробегание по строкам датафрейма
